```python
# what are the most common feature of verginica species

len_sepal=df['sepal_length'].mode()
print("sepal_length",len_sepal)
width_sepal=df['sepal_width'].mode()
print('sepal_width',width_sepal)
length_petal=df['petal_length'].mode().head(1)
print('petal_length',length_petal)
width_petal=df['petal_width'].mode().head(1)
print('petal1_width',width_petal)
```
