# EDA Haberman's Survival Dataset
>Data Description

- Age of patient at time of operation (numerical)
- Patient's year of operation (year - 1900, numerical)
- Number of positive axillary nodes detected (numerical)
- Survival status (class attribute)
- 1 = the patient survived 5 years or longer
- 2 = the patient died within 5 year

```python
# mean

print("means")
print(np.mean(df['Age']))
print(np.mean(df['Operation_year']))
print(np.mean(df["Nb_pos_detected"]))
print(np.mean(df["Survival_status"]))
```

```python
#standard deviations

print("standard deviations")

print(np.std(df['Age']))
print(np.std(df['Operation_year']))
print(np.std(df["Nb_pos_detected"]))
print(np.std(df["Survival_status"]))
```

```python
#median

print("medians")
print(np.median(df['Age']))
print(np.median(df['Operation_year']))
print(np.median(df["Nb_pos_detected"]))
print(np.median(df["Survival_status"]))
```

```python
# quantiles

print("quantiles")
print(np.percentile(df['Age'],np.arange(0,100,25)))
print(np.percentile(df['Operation_year'],np.arange(0,100,25)))
print(np.percentile(df["Nb_pos_detected"],np.arange(0,100,25)))
print(np.percentile(df["Survival_status"],np.arange(0,100,25)))
```

```python
# 90th percentile
print("90th percentile:")
print(np.percentile(df['Age'],90))
print(np.percentile(df['Operation_year'],90))
print(np.percentile(df["Nb_pos_detected"],90))
print(np.percentile(df["Survival_status"],90))
```

```python
# median absolute deviations

from statsmodels import robust
print("MAD")
print(robust.mad(df['Age']))
print(robust.mad(df['Operation_year']))
print(robust.mad(df["Nb_pos_detected"]))
print(robust.mad(df["Survival_status"]))
```

Number of nodes Distribution Distribution

```python
sns.FacetGrid(df,hue="Survival_status",size=6)\
.map(sns.distplot,"Nb_pos_detected")\
.add_legend()
plt.show()
```

```python
# Nb_Pos_detected is imp feature
```

Patient Age CDF and PDF

```python
# compute pdf

counts,bin_edges=np.histogram(df["Age"],bins=10)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)

#compute cdf
cdf=np.cumsum(pdf)
print(cdf)

#plotting pdf nd cdf
plt.plot(bin_edges[1:],pdf) 
plt.plot(bin_edges[1:],cdf)
plt.show()
```

Num of nodes CDF and PDF

```python
count,bin_edges=np.histogram(df["Nb_pos_detected"],bins=10)
pdf=count/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
# 60% of patients had less than 10 nodes
# no patient survived after 5 yrs of surgery who had more than 30 nodes
```
